##### Change Data Type of Sale Price and Market Price from object to numric

      Create New Feature price_diff which is diff between market price and sale price

```python
jio['sale_price']=pd.to_numeric(jio['sale_price'].str.replace(',','').str.strip())
jio['market_price']=pd.to_numeric(jio['market_price'].str.replace(',','').str.strip())
jio['price_diff']=jio['market_price']-jio['sale_price']
```

```python
jio['rating']=jio['rating'].apply(lambda x: round(x,1))
```

#### Products sell at market value

```python
market_price_sell=jio[jio['price_diff']==0.0]
market_price_sell.head()
```

#### Top 10 Prodcuts sell at Market Price by type

```python
top_10_product_sell_market_price=market_price_sell.type.value_counts().sort_values(ascending=False)[:10]
top_10_product_sell_market_price
```

```python
# Plot histogram of category count of top 10 product  where product has same sale price and market price
df=market_price_sell.type.value_counts().sort_values(ascending=False)[:10].reset_index()
df.type
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
df.plot.bar(ax=ax)
ax.set_xticklabels(df.type)
ax.set_xlabel('Category Type')
ax.set_ylabel('Count')
ax.set_title('Top 10 Product Category Sell at market price')
```

```python
top_10=top_10_product_sell_market_price.index
mask=jio['type'].isin(list(top_10))  & (jio['price_diff']!=0.0)
top_10_products_without_sell_diff=jio[mask].type.value_counts().sort_values(ascending=False)
```

```python
top_10_products_without_sell_diff
```

```python
top_10_products_without_sell_diff.index
```

```python
no_product_sell_at_market_price=pd.DataFrame([top_10_product_sell_market_price,top_10_products_without_sell_diff],
             columns=list(top_10),index=['Sale At Market Price','Sale With Profit']).T
```

```python
no_product_sell_at_market_price.sort_values('Sale At Market Price',ascending=False)
```

#### Products with rating above 4.5 and sell at market price

```python
mask=(jio['rating']>4.5) & (jio['price_diff']==0.0)
highest_rated_produt_sell_at_market_price=jio[mask]
highest_rated_produt_sell_at_market_price
```

```python
fig=plt.figure(figsize=(13,12))
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
fig.subplots_adjust(wspace=0,hspace=.6)
highest_rated_produt_sell_at_market_price['category'].value_counts().plot.bar(ax=ax1)
highest_rated_produt_sell_at_market_price['type'].value_counts().plot.bar(ax=ax2)
highest_rated_produt_sell_at_market_price['category'].value_counts().plot.pie(ax=ax3,autopct='%1.1f%%')
ax1.set_xlabel('Category',fontsize=10)
ax2.set_xlabel('Type',fontsize=10)
```
